In this notebook, I will be attempting to compare the performance of 3 of the most cryptocurrencies in the world for the past 3 months, using Python programming. They are:

1. Bitcoin
2. Ethereum
3. Ripple

Bitcoin requires no introduction as it has been around for over a decade, being the oldest cryptocurrency and accepted as a form of payment by many reputable world-wide organisations. Little is known about who developed/ maintains this currency(only that it goes by the pseudonymn Satoshi Nakamoto). 

Ethereum is the second most popular cryptocurrency. It was created by Vitalik Buterin in 2015 and runs on a blockchain-based platform. All contracts/projects/applications that use blockchain could easily be implemented on Ethereum platform.

Ripple is my choice as number 3. The main difference between Ripple and the previous 2 cryptocurrencies is the (de)centralisation approaches. Ripple was created in 2012 with the main objective to increase the speed of International Payment Transfers. It is faster and some would claim cheaper in terms of fees than any other currency in the world in International Transfers, including the more traditional fiat currencies(which require at least a few days.

The biggest downside to transacting in crypto is that it requires a lot of resources(time, elctricity and hardware) having to be deployed. For a cryptocurrency holder, other risks are associated(like market volatility, potential introduction of tighter regulations,etc.)



```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
ticker = ["BTC-USD", "ETH-USD", "XRP-USD"]
```

```python
cryptos = yf.download(ticker, start = "2016-01-01", end = "2020-03-22")
```

```python
# Change the column title names

cryptos.rename(columns={'BTC-USD':'Bitcoin',
                          'ETH-USD':'Ethereum',
                          'XRP-USD':'Ripple'}, 
                 inplace=True)
```

The columns' titles were changed successfully. I will only be using the 'Close' column for the 3 cryptocurrencies( for now).

```python
close = cryptos.loc[:, "Close"].copy()
```

```python
close.head()
```

```python
close.describe()
```

```python
close.plot(figsize = (15, 8), fontsize = 16)
plt.title("Cryptocurrencies Performance Before Normalisation", fontsize = 32)
plt.legend(fontsize = 16)
plt.show()
```

# Normalisation of Time Series to a Base Value of 100

```python
close.head()
```

```python
close.iloc[0, 0]
```

```python
close["Bitcoin"].div(close.iloc[0, 0]).mul(100)
```

```python
close.iloc[0]
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize = (15, 8), fontsize = 16)
plt.title("Cryptocurrencies Performance After Normalisation", fontsize = 32)
plt.legend(fontsize = 16)
plt.show()
```

From the graph above Ethereum shows a significant better performance than the other 2. Surprisingly, Bitcoin shows a dissapointing performance by comparison overall(pretty much flatline).
